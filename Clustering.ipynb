##### 1.What is unsupervised learning in the context of machine learning?

##### 2.How does K-Means clustering algorithm work?

```python
#Ans - It groups data into K clusters by assigning points to the nearest cluster center and updating centers repeatedly.
```

##### 3.Explain the concept of a dendogram in hierarchical clustering.

##### 4.What is the main difference between K-Means and Hierarchical Clustering?

```python
#Ans - Main difference:
#K-Means needs a fixed number of clusters, while Hierarchical Clustering builds a tree and doesn’t need a predefined cluster count.
```

##### 5.What are the advantages of DBSCAN over K-Means?

##### 6.When would you use Silhouette Score in clustering?

```python
#Ans-It is used to measure how well data points fit within their clusters and to choose the best number of clusters.
```

##### 7.What are the limitations of Hierachical Clustering?

##### 8.Why is feature scaling important in clustering algorithms like K-Means?

##### 9.How does DBSCAN identify noise points?

```python
#Ans-Noise in DBSCAN:
#Points that don’t belong to any cluster (not dense enough) are marked as noise.
```

##### 10.Define inertia in the context of K-Means.

```python
#Ans-It is the sum of squared distances between points and their cluster centers; lower is better.
```

##### 11.What is the elbow method in K-Means clustering?

```python
#Ans -It helps find the optimal number of clusters by plotting inertia and looking for a "bend" or elbow. 
```

##### 12.Describe the concept of "density" in DBSCAN.

```python
#Ans-Density in DBSCAN:
#It refers to the number of points within a certain distance (radius) from a point.
```

##### 13.Can hierarchical clustering be used on categorical data?

##### 14.What does a negative Silhouette Score indicate?

##### 15.Explain the term "linkage criteria" in hierarchical clustering.

```python
#Ans - Linkage criteria:
#It defines how the distance between clusters is calculated (e.g., single, complete, average).
```

##### 16.Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?

```python
#Ans-K-Means on varying sizes/densities:
#It assumes equal-sized, equally dense clusters, so it may group data incorrectly.
```

##### 17.What are the core parameters in DBSCAN, and how do they influence clustering?

```python
#Ans-Core parameters in DBSCAN:
#eps (radius) and min_samples (minimum points). They control cluster size and density.
```

##### 18.How does K-Means ++ improve upon standard K-Means initialization?

```python
#Ans-K-Means++ improvement:
#It chooses better initial centers, reducing the chances of poor clustering.
```

##### 19.What is agglomerative clustering?

```python
#Ans-It’s a bottom-up method where each point starts as its own cluster and merges step by step.
```

##### 20.What makes Silhouette Score a better metric than just inertia for model evaluation?

```python
#Ans-Silhouette Score considers both compactness and separation; inertia only measures compactness.
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

##### 21.Generate synthetic data with 4 centers using make_blobs and apply K - Meanse clustering.Visualize using a scatter plot.

```python
from sklearn.datasets import make_blobs
X, _ = make_blobs(n_samples=300, centers=4, random_state=42)
kmeans = KMeans(n_clusters=4, random_state=42)
labels = kmeans.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], color='red')
plt.title("K-Means Clustering")
plt.show()
```

##### 22.Load the iris dataset and use Agglomerative clustering to group the data into 3 clusters. Display the first 10 predicted labels.

```python
from sklearn.datasets import load_iris
from sklearn.cluster import AgglomerativeClustering

iris = load_iris()
X = iris.data
model = AgglomerativeClustering(n_clusters=3)
labels = model.fit_predict(X)

print("First 10 predicted labels:", labels[:10])
```

##### 29.Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart.

```python
from sklearn.datasets import make_blobs
from sklearn.metrics import silhouette_score

X, _ = make_blobs(n_samples=300, centers=4, random_state=42)
scores = []
ks = range(2, 6)

for k in ks:
    labels = KMeans(n_clusters=k, random_state=42).fit_predict(X)
    score = silhouette_score(X, labels)
    scores.append(score)

plt.bar(ks, scores)
plt.title("Silhouette Scores for K=2 to 5")
plt.xlabel("Number of clusters")
plt.ylabel("Score")
plt.show()
```

##### 30.Load the iris dataset and use hierarchical clustering to group data. Plot a dendrogram with average linkage.

```python
from sklearn.datasets import load_iris
from scipy.cluster.hierarchy import dendrogram, linkage

iris = load_iris()
Z = linkage(iris.data, method='average')

plt.figure(figsize=(10, 5))
dendrogram(Z)
plt.title("Dendrogram - Average Linkage")
plt.show()
```

##### 31.Generate synthetic data with overlaping clusters using make_blobs, then apply K - Means and visualize with decision boundaries.

```python
from sklearn.datasets import make_blobs

X, _ = make_blobs(n_samples=300, centers=3, cluster_std=2.0, random_state=42)
kmeans = KMeans(n_clusters=3, random_state=42).fit(X)

# Meshgrid
x_min, x_max = X[:, 0].min()-1, X[:, 0].max()+1
y_min, y_max = X[:, 1].min()-1, X[:, 1].max()+1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                     np.arange(y_min, y_max, 0.1))
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.3)
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, cmap='viridis')
plt.title("K-Means with Decision Boundaries")
plt.show()
```

##### 34.Load the Breast Cancer dataset and compare inertia values for K = 2 to 6 using K - Means. Show results in a line plot.

```python
from sklearn.datasets import load_breast_cancer

X = StandardScaler().fit_transform(load_breast_cancer().data)
inertias = []

for k in range(2, 7):
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(X)
    inertias.append(km.inertia_)

plt.plot(range(2, 7), inertias, marker='o')
plt.title("Inertia vs. K")
plt.xlabel("K")
plt.ylabel("Inertia")
plt.show()
```
