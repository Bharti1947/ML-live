##### 1.What is K-Nearest Neighbors(KNN) and How does it work?

```python
#Ans -KNN is a supervised learning algorithm that finds the K closest data points to a new input and predicts the output based on them. 
#It uses distance (like Euclidean) to find neighbors.
```

##### 2.What is the difference between KNN Classification and KNN Regression?

```python
#Ans-KNN Classification predicts the most common class among neighbors.
#KNN Regression predicts the average value of the neighbors.
```

##### 3.What is the role of the distance metric in KNN?

```python
#Ans - It helps measure how close or far other data points are. 
#KNN uses this to find the "nearest" neighbors for prediction.
```

##### 4.What is the curse of Dimesionality in KNN?

```python
#Ans-When the number of features (dimensions) is too high, distances between points become similar, making KNN less accurate and slower.
```

##### 5.How can we choose the best value of K in KNN?

```python
#Ans -Try different K values using cross-validation, and pick the one that gives the best accuracy or lowest error on validation data. 
```

##### 6.What are KD Tree and Ball Tree in KNN?

```python
#Ans-They are data structures that organize data to make nearest neighbor search faster, especially for large datasets.
```

##### 7.When should you use KD Tree vs.Ball Tree?

```python
#Ans-KD Tree is better for low-dimensional data (less than 20 features).
#Ball Tree is better for high-dimensional or scattered data.
```

##### 8.What are the disadvantages of KNN?

```python
#Ans - Disadvantages of KNN:-
#Slow prediction time on large datasets
#Sensitive to noisy or irrelevant features
#Requires good feature scaling
```

##### 9.How does feature scaling affect KNN?

```python
#Ans - KNN depends on distance, so if features are not scaled, some features can dominate and mislead the model.
```

##### 10.What is PCA(Principal Component Analysis)?

```python
#Ans -PCA is a technique to reduce the number of features in your dataset by keeping only the most important patterns or variations.
```

##### 11.How does PCA work?

```python
#Ans -PCA transforms the data into new features (principal components) by finding directions of maximum variance and projecting the data onto those directions. 
```

##### 12.What is the geometric intuition behind PCA?

```python
#Ans -PCA rotates the data to find the directions (axes) with the most variance and projects the data onto them.
```

##### 13.What is the difference between Feature Selection and Feature Extraction?

```python
#Ans-Feature selection picks existing features; feature extraction creates new ones from original data.
```

##### 14.What are Eigenvalues and Eigenvectors in PCA?

```python
#Ans-Eigenvectors define the direction of new axes; eigenvalues show the importance (variance) along those axes.
```

##### 15.How do you decide the number of components to keep in PCA?

```python
#Ans-Keep enough components to retain a high percentage (e.g., 95%) of the total variance.
```

##### 16.Can PCA be used for classification?

```python
#Ans-PCA itself is not a classifier, but it helps improve classification by reducing dimensions.
```

##### 17.What are the limitations of PCA?

```python
#Ans - It assumes linearity, ignores class labels, and can lose interpretability of features.
```

##### 18.How do KNN and PCA complement each other?

```python
#Ans - PCA reduces dimensions, which can improve KNNâ€™s speed and accuracy by removing noise.
```

##### 19.How does KNN handle missing values in a dataset?

```python
#Ans-KNN can impute missing values by averaging values from the nearest neighbors.
```

##### 20.What are the key differences between PCA and Linear Discriminant Analysis(LDA)?

```python
#Ans-PCA focuses on variance, is unsupervised; LDA focuses on class separation, is supervised.
```

## Practical 

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error
```

##### 21.Train a KNN Classifier on the Iris dataset and print model accuracy.

```python
from sklearn.datasets import load_iris

data = load_iris()
X, y = data.data, data.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = KNeighborsClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
```

##### 22.Train a KNN Regressor on a synthetic dataset and evaluate using Mean Squared Error(MSE).

```python
from sklearn.datasets import make_regression

X, y = make_regression(n_samples=100, n_features=1, noise=10)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = KNeighborsRegressor()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print("MSE:", mean_squared_error(y_test, y_pred))
```

##### 23.Train a KNN Classifier using different distance metrics(Euclidean and Manhattan) and compare accuracy.

```python
knn_euclidean = KNeighborsClassifier(metric='euclidean')
knn_euclidean.fit(X_train, y_train)
acc_euclidean = accuracy_score(y_test, knn_euclidean.predict(X_test))

knn_manhattan = KNeighborsClassifier(metric='manhattan')
knn_manhattan.fit(X_train, y_train)
acc_manhattan = accuracy_score(y_test, knn_manhattan.predict(X_test))

print("Euclidean Accuracy:", acc_euclidean)
print("Manhattan Accuracy:", acc_manhattan)
```

##### 25.Apply Feature Scaling before training a KNN model and compare results with unscaled data.

```python
from sklearn.preprocessing import StandardScaler

# Without Scaling
model_unscaled = KNeighborsClassifier()
model_unscaled.fit(X_train, y_train)
acc_unscaled = accuracy_score(y_test, model_unscaled.predict(X_test))

# With Scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

model_scaled = KNeighborsClassifier()
model_scaled.fit(X_train_scaled, y_train)
acc_scaled = accuracy_score(y_test, model_scaled.predict(X_test_scaled))

print("Unscaled Accuracy:", acc_unscaled)
print("Scaled Accuracy:", acc_scaled)
```

##### 27.Apply PCA before training a KNN Classifier and compare accuracy with and without PCA.

```python
from sklearn.decomposition import PCA

# Without PCA
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
acc_original = accuracy_score(y_test, knn.predict(X_test))

# With PCA
pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)

knn_pca = KNeighborsClassifier()
knn_pca.fit(X_train_pca, y_train)
acc_pca = accuracy_score(y_test, knn_pca.predict(X_test_pca))

print("Accuracy without PCA:", acc_original)
print("Accuracy with PCA:", acc_pca)
```

##### 28.Perform Hyperparameter Tuning on a KNN Classifier using GridSearchCV.

##### 31.Train a KNN classifier using different values of the weights parameter (uniform vs.distance) and compare accuracy.

```python
from sklearn.datasets import load_iris

data = load_iris()
X, y = data.data, data.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

knn_uniform = KNeighborsClassifier(weights='uniform')
knn_uniform.fit(X_train, y_train)
acc_uniform = accuracy_score(y_test, knn_uniform.predict(X_test))

knn_distance = KNeighborsClassifier(weights='distance')
knn_distance.fit(X_train, y_train)
acc_distance = accuracy_score(y_test, knn_distance.predict(X_test))

print("Uniform Weights Accuracy:", acc_uniform)
print("Distance Weights Accuracy:", acc_distance)
```

##### 32.Train a KNN Regressor and analyze the effect of different K values on performance.

```python
k_values = [1, 3, 5, 7, 9]
for k in k_values:
    model = KNeighborsRegressor(n_neighbors=k)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    print(f"K = {k}, MSE = {mse:.2f}")
```

##### 35.Train a KNN Classifier using the KD Tree and Ball Tree algorithms and compare performance.

```python
import time

# KD Tree
kd_model = KNeighborsClassifier(algorithm='kd_tree')
start_kd = time.time()
kd_model.fit(X_train, y_train)
kd_time = time.time() - start_kd
acc_kd = accuracy_score(y_test, kd_model.predict(X_test))

# Ball Tree
ball_model = KNeighborsClassifier(algorithm='ball_tree')
start_ball = time.time()
ball_model.fit(X_train, y_train)
ball_time = time.time() - start_ball
acc_ball = accuracy_score(y_test, ball_model.predict(X_test))

print("KD Tree - Accuracy:", acc_kd, "Time:", round(kd_time, 4), "sec")
print("Ball Tree - Accuracy:", acc_ball, "Time:", round(ball_time, 4), "sec")
```

##### 38.Train a PCA model and analyze the effect of different numbers of components on accuracy.

```python
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.metrics import accuracy_score

# Load data
X, y = load_digits(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

# Try different PCA components
for n in [5, 10, 20, 40]:
    pca = PCA(n_components=n)
    X_train_pca = pca.fit_transform(X_train)
    X_test_pca = pca.transform(X_test)

    knn = KNeighborsClassifier()
    knn.fit(X_train_pca, y_train)
    acc = knn.score(X_test_pca, y_test)
    print(f'Accuracy with {n} components: {acc:.2f}')
```

##### 39.Train a KNN Classifier with different leaf_size values and compare accuracy.

```python
from sklearn.datasets import load_iris
X, y = load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

for leaf in [10, 20, 30, 40, 50]:
    knn = KNeighborsClassifier(leaf_size=leaf)
    knn.fit(X_train, y_train)
    acc = knn.score(X_test, y_test)
    print(f'Accuracy with leaf_size={leaf}: {acc:.2f}')
```

##### 46.Train a PCA model and visualize the data reconstruction error after reducing dimensions.

##### 47.Train a KNN Classifier and visualize the decision boundary.

```python
from sklearn.datasets import make_classification
from sklearn.neighbors import KNeighborsClassifier

X, y = make_classification(n_features=2, n_redundant=0, n_informative=2)
model = KNeighborsClassifier().fit(X, y)

x_min, x_max = X[:, 0].min(), X[:, 0].max()
y_min, y_max = X[:, 1].min(), X[:, 1].max()
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 200),
                     np.linspace(y_min, y_max, 200))
Z = model.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.3)
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.show()
```
